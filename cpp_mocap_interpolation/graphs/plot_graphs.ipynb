```python
import numpy as np
import matplotlib.pyplot as plt
```

# Linear Euler vs. Bezier Euler

```python
# load input motion
infile_path = f"./131_04-dance.amc"
data_input = []
counter = 0
with open(infile_path, 'r') as infile:
    for line in infile:
        values = line.split()
        if not values:
            continue
            
        if values[0] == "lfemur" and len(values) == 4:
            counter += 1
            if counter >= 600 and counter <= 800:
                data_input.append(float(values[1]))
            elif(counter > 800):
                break

# load data for linear Euler interpolation 
infile_path = f"./131-linearEuler.amc"
data_le = []
counter = 0
with open(infile_path, 'r') as infile:
    for line in infile:
        values = line.split()
        if not values:
            continue
            
        if values[0] == "lfemur" and len(values) == 4:
            counter += 1
            if counter >= 600 and counter <= 800:
                data_le.append(float(values[1]))
            elif(counter > 800):
                break
            
# load data for Bezier Euler interpolation 
infile_path = f"./131-bezierEuler.amc"
data_be = []
counter = 0
with open(infile_path, 'r') as infile:
    for line in infile:
        values = line.split()
        if not values:
            continue
            
        if values[0] == "lfemur" and len(values) == 4:
            counter += 1
            if counter >= 600 and counter <= 800:
                data_be.append(float(values[1]))
            elif(counter > 800):
                break
```

```python
plt.figure(figsize=(10, 8))
plt.plot(range(600, 600+len(data_input)), data_input, label="Input")
plt.plot(range(600, 600+len(data_le)), data_le, label="Linear Euler")
plt.plot(range(600, 600+len(data_be)), data_be, label="Bezier Euler")
plt.xlabel("Frame Number")
plt.ylabel("RotX Euler Angle (degrees)")
title = "Linear Euler vs Bezier Euler" 
plt.title(title)
plt.legend()
plt.grid()
plt.savefig("LinearEuler_vs_BezierEuler.png")
plt.show()
```
